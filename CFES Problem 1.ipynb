```python
#from urllib.request import urlretrieve as retrieve
import requests
```

```python
## Create two dictionaries of 'url' addresses per state
## - Each dictionary key will hold the name and extension we want our filed to be saved under
    ## - Save the dowloaded file with State name and .html extension
## - Each dictonary value will hold the direct download link for each xls file
    ## - NOTE the file is acutally HTML saved as xls format 
        ## - Dictionary 1 will hold the Public District info within each State
        ## - Dictionary 2 will hold the Public Schools info within each State

url = { 'Alabama.html' : 'https://nces.ed.gov/tempfiles/excelcreator/ncesdata_5107175B.xls',
        #'Alaska.html' : 'https://nces.ed.gov/tempfiles/excelcreator/ncesdata_C4078CFB.xls',
        'Arizona.html' : 'https://nces.ed.gov/tempfiles/excelcreator/ncesdata_C9C543D0.xls',
        #'Arkansas.html' : 'https://nces.ed.gov/tempfiles/excelcreator/ncesdata_3C3A77A2.xls',
        'California.html' : 'https://nces.ed.gov/tempfiles/excelcreator/ncesdata_911234A6.xls'}
```

```python
## Download and rename files
import pandas as pd

for i, address in url.items():
    g = requests.get(address)
    open(i, 'wb').write(g.content)
```

```python
## Convert HTML to Pandas DataFrame and save under new dictonary 'states'
import pandas as pd 

states = {}

for i in url:
    name = i[:-5]
    df = pd.read_html(i)[0]  # Load HTML into a Pandas DataFrame
    df.columns = ['NCES District ID', 
                  'State District ID', 
                  'District Name', 
                  'County Name',
                  'Street Address', 
                  'City', 'State', 
                  'ZIP', 
                  'ZIP 4-digit', 
                  'Phone', 
                  'Students',
                  'Teachers', 
                  'Schools', 
                  'Locale Code', 
                  'Locale', 
                  'Student Teacher Ratio', 
                  'Type']  # Assign New Header Column (Remove '*' in original header)
    clean_data = df[6:].reset_index(drop=True) # Slice DataFrame and Reset Index         
    temp_dict = { name : clean_data}
    states.update(temp_dict)

```
